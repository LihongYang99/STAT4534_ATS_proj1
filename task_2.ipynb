2-1 Perform and plot a square-root transformation of the time series data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("/Users/lihong/Desktop/ATS_proj1/AirPassengers.csv")

# Convert fractional year to datetime
def year_fraction_to_datetime(year_float):
    year = int(year_float)
    fraction = year_float - year
    month = int(round(fraction * 12)) + 1
    if month > 12:   # fix rounding overflow
        year += 1
        month = 12
    return pd.Timestamp(year=year, month=month, day=1)

df["Date"] = df["Time"].apply(year_fraction_to_datetime)
ts = df.set_index("Date")["AirPassengers"].asfreq("MS")  # monthly frequency

# ==== Square-root transformation ====
ts_sqrt = np.sqrt(ts)

# ==== Plot raw vs sqrt transform ====
fig, ax = plt.subplots(2, 1, figsize=(12,7), sharex=True)

ax[0].plot(ts.index, ts.values, marker="o", linewidth=1, label="Raw")
ax[0].set_title("Raw Time Series (AirPassengers)")
ax[0].set_ylabel("Passengers")
ax[0].grid(True)

ax[1].plot(ts_sqrt.index, ts_sqrt.values, marker="o", linewidth=1, color="darkorange", label="Sqrt Transform")
ax[1].set_title("Square-root Transformed Time Series")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("sqrt(Passengers)")
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

2-2 Perform and plot a logarithm transformation of the time series data.


```python
# ==== Convert fractional year to datetime ====
def year_fraction_to_datetime(year_float):
    year = int(year_float)
    fraction = year_float - year
    month = int(round(fraction * 12)) + 1
    if month > 12:
        year += 1
        month = 12
    return pd.Timestamp(year=year, month=month, day=1)

df["Date"] = df["Time"].apply(year_fraction_to_datetime)
ts = df.set_index("Date")["AirPassengers"].asfreq("MS")

# ==== Log10 transform ====
ts_log10 = np.log10(ts)

# ==== Plot raw vs log10 ====
fig, ax = plt.subplots(2, 1, figsize=(12,7), sharex=True)

ax[0].plot(ts.index, ts.values, marker="o", linewidth=1, label="Raw")
ax[0].set_title("Raw Time Series (AirPassengers)")
ax[0].set_ylabel("Passengers")
ax[0].grid(True)

ax[1].plot(ts_log10.index, ts_log10.values, marker="o", linewidth=1, color="purple", label="log10 Transform")
ax[1].set_title("Log10 Transformed Time Series")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("log10(Passengers)")
ax[1].grid(True)

plt.tight_layout()
plt.show()

```
